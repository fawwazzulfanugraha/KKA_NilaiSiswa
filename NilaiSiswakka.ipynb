```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# nomor 11-21
data = pd.read_csv('nilai_siswa.csv', sep=';')
data.info()
data.head()
data.describe()

# nomor 22-25
print("Rata-rata:", data['Nilai'].mean())
print("Median:", data['Nilai'].median())
print("Modus:", data['Nilai'].mode()[0])
```

26-29. Menampilkan nilai mata pelajaran Matematika, Bahasa Inggris, Bahasa Indonesia, Produktif)

```
# matematika
matematika = data[data['Matpel'] == 'Matematika']
print("Nilai siswa untuk mata pelajaran Matematika:")
print(matematika)

# Bahasa Inggris
bahasa_inggris = data[data['Matpel'] == 'Bahasa Inggris']
print("\nNilai siswa untuk mata pelajaran Bahasa Inggris:")
display(bahasa_inggris)

# Bahasa Indonesia
bahasa_indonesia = data[data['Matpel'] == 'Bahasa Indonesia']
print("\nNilai siswa untuk mata pelajaran Bahasa Indonesia:")
display(bahasa_indonesia)

# Produktif
produktif = data[data['Matpel'] == 'Produktif']
print("\nNilai siswa untuk mata pelajaran Produktif:")
display(produktif)
```

30. Menampilkan nilai maksimum dan minimum per mata pelajaran

```
max_min_nilai = data.groupby('Matpel')['Nilai'].agg(['max','min'])
print("Nilai Maksimum dan Minimum per Mata Pelajaran:")
display(max_min_nilai)
```

32-38. Membuat grafik batang (bar chart) rata-rata nilai per mata pelajaran


```
rata = data.groupby('Matpel')['Nilai'].mean()

plt.figure(figsize=(10, 6))
rata.plot(kind='bar', color='skyblue')
plt.title('Rata-Rata Nilai per Mata Pelajaran')
plt.xlabel('Mata Pelajaran')
plt.ylabel('Nilai Rata-Rata')
plt.xticks(rotation=45, ha='right') # Memutar label X agar mudah dibaca
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

39-42. Membuat diagram boxplot untuk melihat sebaran nilai

```
plt.figure(figsize=(10, 6))
sns.boxplot(x='Matpel', y='Nilai', data=data, palette='viridis')
plt.title('Sebaran Nilai per Mata Pelajaran')
plt.xlabel('Mata Pelajaran')
plt.ylabel('Nilai')
plt.xticks(rotation=45, ha='right') # Memutar label X agar mudah dibaca
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
